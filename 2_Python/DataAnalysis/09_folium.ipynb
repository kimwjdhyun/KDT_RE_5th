## 09. Folium
- leaflet.js 기반으로 만들어진 Python 지도 시각화 라이브러리

```python
import folium
```

```python
# 기본 지도 생성
m = folium.Map(location = [37.61123631026529, 126.91723117419909], zoom_start = 18)

# 기본 마커
folium.Marker([37.61123631026529, 126.91723117419909],
              popup="SubWay",
              tooltip="구산역",
              icon=folium.Icon(color="black", icon="fa-solid fa-bus-simple", prefix="fa-solid")
              ).add_to(m)
m

# 클릭한 곳에 마커 추가
# folium.ClickForMarker()
m.add_child(folium.ClickForMarker(popup="내가 클릭한 곳"))

# 클릭한 곳의 위도와 경도 표시
# Folium.LatLngPopup ()
m.add_child(folium.LatLngPopup())

# 지도에 원을 표시(원형 마커)
# Folium.Circle()
folium.CircleMarker(
    [37.6104, 126.9128],
    radius=100,
    color = "#020714",
    fill_color = "#1cffe3",
    popup="CircleMarker",
    tooltip="tooltip"
).add_to(m)
m
```

```python
# 택시승차대 리더님 해설
import pandas as pd
df = pd.read_csv("./dataset/택시승차대 현황.csv", encoding="cp949")
# df

taxi =  df.loc[:, ["위치명", "위도", "경도"]]
latitude = taxi["위도"].mean()
longitude = taxi["경도"].mean()
m = folium.Map([latitude, longitude], zoom_start=13)

for idx, row in taxi.iterrows():
  # print(idx, row["위치명"], row["위도"], row["경도"])
  folium.Marker(
    [row["위도"], row["경도"]],
    tooltip=row["위치명"],
    icon=folium.Icon(color="orange", icon="fa-solid fa-taxi", prefix="fa-solid")
  ).add_to(m)
  
m
```
