### Series
- 인덱스(라벨)가 붙은 1차원 배열
- 인덱스 이름을 직접 지정할 수 있음

```python
# 시리즈 생성
# 1. 리스트로 생성
data = [10, 20, 30, 40]
s = pd.Series(data)
print(s)

# 0    10
# 1    20
# 2    30
# 3    40
# dtype: int64
```

```python
# 2. 인덱스를 직접 지정
s = pd.Series(data, index=["a", "b", "c", "d"])
print(s)

# a    10
# b    20
# c    30
# d    40
# dtype: int64
```

```python
# 3. 딕셔너리로 생성 
data = {'서울' : 100, '부산' : 200, '대구' : 300, '제주' : 400}
s = pd.Series(data)
print(s)

# 서울    100
# 부산    200
# 대구    300
# 제주    400
# dtype: int64
```

```python
# 4. 하나의 값으로 시리즈를 생성
s = pd.Series(7, index=["가", "나", "다"])
print(s)

# 가    7
# 나    7
# 다    7
# dtype: int64
```

```python
# 시리즈의 주요 속성
data = [10, 20, 30, 40]
s = pd.Series(data, index=['a', 'b', 'c', 'd'], name='test data')
print("values", s.values) # 실제 데이터 배열    values [10 20 30 40]
print("index", s.index)   # 인덱스 객체         index Index(['a', 'b', 'c', 'd'], dtype='object')
print("dtype", s.dtype)   # 실제 데이터의 타입  dtype int64
print("size", s.size)     # 데이터의 개수       size 4
print("shape", s.shape)   # 배열의 형태         shape (4,)
print("name", s.name)     # 시리즈의 이름       name test data
```

```python
# 시리즈의 연산
s1 = pd.Series([10, 20, 30], index=["a", "b", "c"])
s2 = pd.Series([5, 6, 7], index=["c", "b", "d"])

result = s1 + s2
print(result)

# a     NaN
# b    26.0
# c    35.0
# d     NaN
# dtype: float64
```

```python
# 브로드 캐스팅
s = pd.Series([5, 10, 15], index=['x', 'y', 'z'])

print(s + 3)
# x     8
# y    13
# z    18
# dtype: int64
print(s * 2)
# x    10
# y    20
# z    30
# dtype: int64
```

```python
# 불리언 연산 및 인덱싱
s = pd.Series([50, 60, 70, 80, 90, 100])

mask = s > 70
print(mask)
# 0    False
# 1    False
# 2    False
# 3     True
# 4     True
# 5     True
# dtype: bool
print(s[mask])
# 3     80
# 4     90
# 5    100
# dtype: int64
```

### iloc / loc

```python
# iloc
# integer location
# 정수 위치 기반의 인덱싱, 슬라이싱
# NumPy의 슬라이싱과 유사하게 동작함

# 단일 행 / 열 선택
df.iloc[0] # 첫 번째 행
df.iloc[:, 0:2]
# 이름    홍길동
# 나이     23
# 직업     학생
# Name: 0, dtype: object

df.iloc[:, 0:2] # 두 번째 열
#   이름	나이
# 0	홍길동	23
# 1	이순신	35
# 2	김유신	31
# 3	강감찬	40
# 4	장보고	28
# 5	이방원	34
```

```python
# 여러 행 / 열 동시 선택
df.iloc[1:4]
#   이름	나이	직업
# 1	이순신	35	    군인
# 2	김유신	31	    장군
# 3	강감찬	40	    장군

df.iloc[:, 0:2]
# 	이름	나이
# 0	홍길동	23
# 1	이순신	35
# 2	김유신	31
# 3	강감찬	40
# 4	장보고	28
# 5	이방원	34

df.iloc[1:4, 0:2]
# 	이름	나이
# 1	이순신	35
# 2	김유신	31
# 3	강감찬	40
```

```python
# Fancy Indexing

df.iloc[[0, 2, 4]]
# 	이름	나이	직업
# 0	홍길동	23	    학생
# 2	김유신	31	    장군
# 4	장보고	28	    상인

df.iloc[:, [1, 2]]

#      나이	        직업
# 0	    23	        학생
# 1	    35	        군인
# 2	    31	        장군
# 3	    40	        장군
# 4	    28	        상인
# 5	    34	        왕자

df.iloc[[0, 2, 4], [1, 2]]
# 	나이	직업
# 0	23	    학생
# 2	31	    장군
# 4	28	    상인
```

```python
# 음수 인덱스 슬라이싱
df.iloc[-1]
# 이름    이방원
# 나이     34
# 직업     왕자
# Name: 5, dtype: object

df.iloc[:, -2:]
# 	나이	직업
# 0	23	    학생
# 1	35	    군인
# 2	31	    장군
# 3	40	    장군
# 4	28	    상인
# 5	34	    왕자
```

```python
# loc
# location
# 라벨의 이름 기준으로 인덱싱 / 슬라이싱
# 단일 행 / 열 선택
# 시작과 끝을 모두 포함
# 음수 인덱스는 사용 x

df.loc[0]
# 이름    홍길동
# 나이     23
# 직업     학생
# Name: 0, dtype: object

df.loc[:, "이름"]
# 0    홍길동
# 1    이순신
# 2    김유신
# 3    강감찬
# 4    장보고
# 5    이방원
# Name: 이름, dtype: object

# 여러 행 / 열 선택

df.loc[2:4, ["이름", "직업"]]
#     이름	직업
# 2	김유신	장군
# 3	강감찬	장군
# 4	장보고	상인

df.loc[2:4, "이름":"직업"]

#   이름	    나이	    직업
# 2	김유신	    31	        장군
# 3	강감찬	    40	        장군
# 4	장보고	    28	        상인
```

#### 결측값 처리
- 결측값(Missing Value, NA, Null) :데이터에 값이 기록되지 않은 상태
> Pandas에서는 보통 NaN(Not a Numver, float), None(object), Na 등으로 처리

```python
data = {
    "이름": ["서준", "하은", "민준", "서연", "이안", "지민"],
    "나이": [22, 28, np.nan, 31, 27, 24],
    "점수": [89, np.nan, 83, 90, 88, 93]
}
df = pd.DataFrame(data)


# 	이름	나이	점수
# 0	서준	22.0	89.0
# 1	하은	28.0	NaN
# 2	민준	NaN	    83.0
# 3	서연	31.0	90.0
# 4	이안	27.0	88.0
# 5	지민	24.0	93.0
```

```python
# 결측값 탐지
# isnull : 결측값이 맞으면 True 아니면 False
df.isnull()
# 	이름	나이	점수
# 0	False	False	False
# 1	False	False	True
# 2	False	True	False
# 3	False	False	False
# 4	False	False	False
# 5	False	False	False

# 각 컬럼의 결측값의 수 계산
df.isnull().sum()
# 이름    0
# 나이    1
# 점수    1
# dtype: int64

# 데이터의 전체 결측값 수 계산
df.isnull().sum().sum()     # np.int64(2)
```

```python
# notnull() : 결측값이 아니면 True 맞으면 False
df.notnull()
    # 이름	나이	점수
# 0	True	True	True
# 1	True	True	False
# 2	True	False	True
# 3	True	True	True
# 4	True	True	True
# 5	True	True	True
```

```python
# 결측값 제거 : dropna()
# 결측값이 있는 행을 삭제
df2 = df.dropna()
df2

#   이름	나이	점수
# 0	서준	22.0	89.0
# 3	서연	31.0	90.0
# 4	이안	27.0	88.0
# 5	지민	24.0	93.0
```

```python
# 결측값이 있는 열을 삭제
df3 = df.dropna(axis=1)
df3

#   이름
# 0	서준
# 1	하은
# 2	민준
# 3	서연
# 4	이안
# 5	지민
```

```python
# 결측값 대체(채우기) : fillna()
# 결측값을 특정 값으로 대체
df4 = df.fillna(0)
df4

# 	이름	나이	점수
# 0	서준	22.0	89.0
# 1	하은	28.0	0.0
# 2	민준	0.0	    83.0
# 3	서연	31.0	90.0
# 4	이안	27.0	88.0
# 5	지민	24.0	93.0

avg_age = df["나이"].mean()
avg_age     # np.float64(26.4)

df["나이"] = df["나이"].fillna(avg_age)
df

# 	이름	나이	점수
# 0	서준	22.0	89.0
# 1	하은	28.0	NaN
# 2	민준	26.4	83.0    # 평균값 추가!
# 3	서연	31.0	90.0
# 4	이안	27.0	88.0
# 5	지민	24.0	93.0
```

```python
# 이전 값으로 결측값 채우기 (foward fill)
df5 = df.ffill()
df5

# 	이름	나이	점수
# 0	서준	22.0	89.0
# 1	하은	28.0	89.0  0번 인덱스 값으로 채움(이전 값)
# 2	민준	26.4	83.0
# 3	서연	31.0	90.0
# 4	이안	27.0	88.0
# 5	지민	24.0	93.0
```

```python
# 다음 값으로 결측값 채우기 (backward fill)
df6 = df.bfill()
df6


#     이름	나이	점수
# 0	서준	22.0	89.0
# 1	하은	28.0	83.0    2번 인덱스 값으로 변경 (다음 값)
# 2	민준	26.4	83.0 
# 3	서연	31.0	90.0
# 4	이안	27.0	88.0
# 5	지민	24.0	93.0
```

```python
# 실습4. 통계 함수 · 결측값 처리 연습(1)
data = {
"도시": ["서울", "부산", "광주", "대구", np.nan, "춘천"],
"미세먼지": [45, 51, np.nan, 38, 49 , np.nan],
"초미세먼지": [20, np.nan, 17, 18, 22, 19],
"강수량": [0.0, 2.5, 1.0, np.nan, 3.1, 0.0]
}
df = pd.DataFrame(data)
#     도시	미세먼지	초미세먼지	강수량
# 0	서울	45.0	    20.0	    0.0
# 1	부산	51.0	    NaN	        2.5
# 2	광주	NaN	        17.0	    1.0
# 3	대구	38.0	    18.0	    NaN
# 4	NaN	    49.0	    22.0	    3.1
# 5	춘천	NaN	        19.0	    0.0
```

```python
# 실습4 . 통계 함수 · 결측값 처리 연습(2)
# 1. ‘미세먼지’ 컬럼의 평균과 중앙값을 구하세요.
print(f"문제 1 . \n 평균 : {df["미세먼지"].mean()} \n 중앙값 : {df["미세먼지"].median()}")

# 2. ‘초미세먼지’ 컬럼의 최댓값과 최솟값을 구하세요.
print(f"\n문제 2 . \n최댓값 : {df["초미세먼지"].max()} \n최솟값 : {df["초미세먼지"].min()}")

# 3. 각 컬럼별 결측값 개수를 구하세요.
print("\n문제 3 . \n", df.isnull().sum())
```

```python
# 4. 결측값이 하나라도 있는 행을 모두 삭제한 뒤, 남은 데이터의 ‘초미세먼지’ 평균을 구하세요.
df1 = df.dropna(axis=0, how="any")
print("문제 4 . \n","초미세먼지 평균 : ", df1["초미세먼지"].mean())

# 5. 결측값을 모두 0으로 채운 뒤, ‘미세먼지’와 ‘초미세먼지’의 합계를 각각 구하세요.
df2 = df.fillna(0)
print(f"\n문제 5 . \n미세먼지 합 :{df2["미세먼지"].sum()} \n초미세먼지 합 : {df["초미세먼지"].sum()}")

# 6. ‘미세먼지’ 컬럼의 결측값을 평균값으로 채운 뒤, 그 표준편차를 구하세요
df3 = df.fillna(df["미세먼지"].mean())
df3
print("\n문제 6 . \n", df3)
print("표준편차 : ", df3["미세먼지"].std())
```
