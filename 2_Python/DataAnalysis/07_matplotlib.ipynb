## 07. Matplotlib
- 파이썬에서 가장 널리 사용되는 데이터 시각화 라이브러리!

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)

plt.rcParams['font.family'] = 'malgun gothic' # '맑은 고딕'이 설치되어 있을 경우
plt.rcParams['axes.unicode_minus'] = False    # 마이너스(-) 부호 깨짐 방지
```

### 그래프 그리기

```python
# 그래프 그리기 : pyplot.plot()
x = [1, 2, 3, 4, 5]
y = [10, 17, 13, 19, 24]

# 선 그래프
plt.plot(x, y)

# 이미지를 파일로 저장 : pyplot.savefig("파일 저장명")
plt.savefig("my_graph.png")
# <Figure size 640x480 with 0 Axes>

# 콘솔에 이미지 그리기 : pyplot.show()
plt.show()
```

### 그래프 구성요소(1)
- figure : 여러 그래프가 그려지는 최상위 컨테이너
- axes : 실제 그래프가 그려지는 공간
- axis : 그래프의 축

```python
# subplot()
# - 여러 개의 그래프를 그릴 때 사용
# - figure, axes 동시 반환
# fig, axs = plt.subplot(행, 열)
fig, axs = plt.subplots() # 하나의 figure와 하나의 axes를 만듦
```

```python
# 1행 2열의 axes 생성
fig, axs = plt.subplots(1, 2)

axs[0].plot([1, 2, 3], [1, 2, 3])
axs[1].plot([1, 2, 3,], [1, 4, 9])

plt.show()
```

```python
# 2행 2열의 axes 생성
fig, axs = plt.subplots(2, 2)
a = [1, 2, 3]
b = [1, 4, 9]


fig.suptitle("여러 함수 그래프")    # 함수 제목

axs[0, 0].plot(a, a)
axs[0, 0].set_title('1번 그래프')

axs[0, 1].plot(a, b)
axs[0, 1].set_title('2번 그래프')

axs[1, 0].plot(a, [3, 2, 1])
axs[1, 0].set_title('3번 그래프')

axs[1, 1].plot(a, [9, 4, 1])
axs[1, 1].set_title('4번 그래프')

plt.tight_layout()      # 그래프 간격 자동 조정

plt.show()
```

### 그래프의 구성요소(2)
- 제목, 레이블, 범례, 그리드

```python
# title : 제목
plt.plot([1, 2, 3, 4, 5], [1, 4 ,9, 16, 25], label= '제곱 그래프', linestyle= '--', color= 'c', marker='o')
plt.title('제곱 함수 그래프', color= 'k', fontsize= 20)

# xlabel, ylabel
plt.xlabel("x 값")
plt.ylabel("y 값(x의 제곱)")

# 범례
plt.legend()

# 격자
plt.grid(linestyle= '--')

plt.show()
```

```python
# 하나의 axes에 여러 개의 그래프 그리기
x = np.arange(0, 10)
y1 = 2 * x + 1
y2 = 3 * x - 5

plt.plot(x, y1, marker='o', label= 'y = 2x + 1', color= 'c')
plt.plot(x, y2, marker='^', label= 'y = 3x - 5', color= 'm')

plt.title('두 개의 선 그래프 비교')
plt.xlabel("x 값")
plt.ylabel('y 값')
plt.legend()
plt.grid(linestyle= ':', alpha=0.5)

plt.show()
```

```python
# 실습1. 선 그래프 그리기
# 아래 데이터를 이용해서 예시와 같이 선 그래프를 그려보세요.
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_2019 = [100, 120, 140, 110, 130, 150, 160, 170, 180, 200, 190, 210]
sales_2020 = [90, 110, 130, 120, 140, 160, 170, 160, 150, 180, 200, 190]

plt.plot(months, sales_2019, label=2019, color='b')
plt.plot(months, sales_2020, label= 2020, color= 'r')

plt.title('Monthly Sales Comparison (2019 - 2020)')
plt.xlabel("Month", labelpad= 15)
plt.ylabel('Sales', labelpad= 15)
plt.legend(title="판매 연도", loc="lower right")

plt.show()
```

```python
# 막대 그래프(Bar Chart)

labels = ['A', 'B', 'C', 'D']
values = [10, 24, 15, 32]

plt.bar(labels, values)

plt.show()
```

```python
# 수평 막대 그래프 : plt.barh

plt.barh(labels, values)

plt.show()
```

```python
# 여러 막대 그래프
labels = ["1분기", "2분기", "3분기", "4분기"]
men_means = [20, 34, 30, 35]
women_means = [25, 32, 34, 40]

x = np.arange(len(labels))      # x로 기본 위치를 설정해줌

fig, axs = plt.subplots()
width = 0.4                     # 막대 너비를 설정해줌

axs.bar(x - width/2, men_means, width, label= "남성")       # width/2로 두 객체의 위치 선택
axs.bar(x + width/2, women_means, width, label="여성")

axs.set_xticks(x, labels)       # 숫자 대신 문자열을 표시
axs.legend()

plt.show()
```

```python
# 실습2. 막대 그래프 그리기
# 아래 데이터를 바탕으로 예시와 같이 막대 그래프를 그려보세요.
categories = ['Category1','Category2','Category3','Category4','Category5']
data = [20, 35, 15, 27, 45]

x = np.arange(len(categories))

fig, axs = plt.subplots()
axs.bar(x, data, width=0.5, label="Data", color='m')
plt.title("Bar Chart")
plt.xlabel("Categories", labelpad= 15)
plt.ylabel('Values', labelpad= 15)
plt.xticks(rotation=45) # 이름이 긴 경우 각도를 바꿀수있음
plt.yticks(rotation=45)
plt.grid(linestyle=':')
axs.set_xticks(x, categories)
plt.ylim(0, 50)         # y축을 0 ~50으로 고정
axs.legend()

plt.show()
```

```python
# 실습 3. 히스토그램 그리기
# 임의로 1 ~6 사이의 주사위 눈 100개를 생성해서 각 숫자가 얼마나 자주 나왔는지 히스토그램을 그려보세요.
data = np.random.randint(1, 7, size = 100)

plt.hist(data, bins=np.arange(1, 8) - 0.5, edgecolor='k', rwidth=0.8, color='skyblue') # rwidth : 그래프 사이를 띄울 때
plt.title("주사위 눈 빈도 히스토그램")
plt.xticks(range(1, 7))
plt.xlabel("주사위 눈", labelpad= 15)
plt.ylabel('빈도', labelpad= 15)

plt.show()
```

```python
# 산점도(Scatter Plot) 그리기
np.random.seed(2)

x = np.linspace(2, 30, 20)
y = np.linspace(10, 40, 20)
y_noisy = y + np.random.uniform(-7, 7, 20)

plt.scatter(x, y_noisy, color='k', marker='^')
plt.grid(linestyle=':')
plt.show()
```

```python
# 실습4 . 산점도 그리기
# x값은 0부터 10까지 30 개의 균등한 간격으로 구성된 수열을 만들고,
# y값은 y = 3x + 2 + 난수(정규분포, 평균0, 표준편차3)로 생성하여 산점도를 그려보세요.

np.random.seed(10)

x = np.linspace(0, 10, 30)
y = 3 * x + 2 + np.random.normal(0, 3, size = 30)

plt.scatter(x, y, color='purple', edgecolors='b', marker=',')
plt.title("노이즈가 추가된 산점도")
plt.xlabel("x 값", labelpad= 15)
plt.ylabel('y 값', labelpad= 15)
plt.grid(True, linestyle= ':', alpha=0.3)

plt.show()
```

```python
# 파이 차트 (Pie Chart)그리기
labels = ["Apple", "Banana", "Mango", "Blueberry"]
sizes = [15, 30, 45, 10]
explode = (0.1,0.1,0,0)

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels,explode= explode, shadow=True, autopct='%1.1f%%', startangle=90,
        colors=["#020714", "#951C35", "#18D681", "#A65AE1"])
plt.show()
```

```python
# 실습1. 파이 차트 그리기
# 아래와 데이터를 이용해서 예시와 같이 파이 차트를 그려보세요!
np.random.seed(100)
labels = ['Watermelon', 'Grape', 'Applemango', 'Strawberry']
sizes = np.random.randint(1, 100, size=len(labels))
sizes = sizes / sizes.sum() * 100
explode = (0.1, 0, 0.1, 0)

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90,
        colors=["#0ED772", "#890A5D", "#E3DC0F", "#EF0505"])
plt.title('과일 파이 차트(랜덤)')

plt.show()
```

```python
plt.style.use('_mpl-gallery-nogrid')


# make data
x = [1, 2, 3, 4]
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))

# plot
fig, ax = plt.subplots()
ax.pie(x, colors=colors, radius=3, center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
