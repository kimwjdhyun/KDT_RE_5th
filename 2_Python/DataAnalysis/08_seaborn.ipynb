```python
# 스타일 설정
sns.set_style("ticks") # 스타일 테마 설정
sns.set_context("notebook") # 문맥에 따라 스타일 크기 조정
sns.set_palette("pastel")

plt.rcParams['font.family'] = 'malgun gothic' # '맑은 고딕'이 설치되어 있을 경우
plt.rcParams['axes.unicode_minus'] = False    # 마이너스(-) 부호 깨짐 방지
```

```python
# 범주형 데이터 그래프
# countplot
palette = sns.color_palette("pastel")
tips = sns.load_dataset("tips")
sns.countplot(data = tips, x = 'day', hue='sex', palette=palette)
plt.title("요일별 방문자 수")
plt.grid(linestyle=":")
plt.show()
```

```python
# barplot
palette = sns.color_palette("colorblind")
sns.barplot(data = tips, x="day", y = "tip", palette=palette)
plt.title("요일별 평균 팁 금액")
plt.show()
```

```python
# displot() : 히스토그램 + 커널 밀도 (kde)
# kde = True (kde 그래프를 그려줌)
sns.displot(data = tips, x = "total_bill", bins=20, kde=True)
plt.title("Total Bill")
plt.show()
```

```python
# scatterplot() : 산점도

sns.scatterplot(data = tips, x = "total_bill", y = "tip", hue = "sex", )
plt.title()
```

```python
# 실습 2. seaborn 그래프 그리기
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

palette = sns.color_palette("viridis")
titanic = sns.load_dataset("titanic")
plt.rcParams['font.family'] = 'malgun gothic' # '맑은 고딕'이 설치되어 있을 경우
plt.rcParams['axes.unicode_minus'] = False    # 마이너스(-) 부호 깨짐 방지

sns.countplot(data = titanic, x = "pclass", hue='sex', palette=palette)
plt.title("총 승객 등급")
plt.xlabel("승객 등급", labelpad=15)
plt.ylabel("승객 인원 수", labelpad=15)

plt.show()
```

```python
# 실습 2. seaborn 그래프 그리기
sns.barplot(data=titanic, x = "pclass", y = "survived", hue='who', palette=palette)

plt.title("타이타닉 생존자")
plt.xlabel("남, 녀, 아동", labelpad=15)
plt.ylabel("생존 여부", labelpad=15)

plt.show()
```

```python
# 실습 2. seaborn 그래프 그리기
tit = sns.displot(data=titanic, x='age', kde=True, bins=10, 
            hue='survived', palette=['gray', 'green'], col='who')

tit.fig.suptitle("타이타닉 승객의 연령 분포 및 생존 현황", fontsize=16, y=1.02)
tit.set_titles("{col_name}")
tit.set_axis_labels("나이", "인원 수")
plt.tight_layout()
plt.show()
```

```python
# heatmap() - 히트맵
# - 2차원 행렬을 색상을 통해 시각화 하는 방법(특히, 변수 간 상관계수 행렬을 시각화할 때 자주 사용됨)
penguins = sns.load_dataset("penguins")
print(penguins.head(10))

# 상관계수 행렬
# - 상관계수 : 두 변수 사이에 직선 관계가 얼마나 강한지를 숫자로 표현한 값
# - heatmap()에서 색으로 표시되는 숫자가 상관계수
# 1 : 양의 상관관계     (하나가 증가하면 다른 것도 같은 비율로 증가)
# ex) 키와 몸무게
# -1 : 음의 상관관계    (하나가 증가하면 다른 변수는 감소)
# ex) 공부 시간과 실수를 하는 개수

corr = penguins.corr(numeric_only=True)
sns.heatmap(corr, annot=True, fmt=".2f", cmap="inferno")
plt.show()
```

```python
# jointplot() : scatter + histogram/kde
sns.jointplot(data = penguins, x = 'flipper_length_mm', y = "body_mass_g", kind = "reg")
plt.show()
```

```python
# 실습 3. seaborn 그래프 그리기
sns.jointplot(data=penguins, x="bill_length_mm", y='flipper_length_mm',
              kind='scatter', palette=['r', 'g', 'b'], hue='species')

plt.suptitle('펭귄 종별 부리 길이와 물갈퀴 길이의 관계')
plt.xlabel('부리 길이 (mm)', fontsize=11, labelpad=15)
plt.ylabel('물갈퀴 길이 (mm)', fontsize=11, labelpad=15)
plt.grid(linestyle=':')
plt.legend(fontsize=8, markerscale=1, title_fontsize=9, loc='lower right')
plt.tight_layout()
plt.show()
```
