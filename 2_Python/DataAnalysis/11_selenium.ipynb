## Seleneium
- 웹 자동화를 위한 라이브러리

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service

import time
```

```python
# Chrome 브라우저 실행
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# 원하는 페이지로 이동
driver.get("https://www.google.com")    # 구글로 이동
time.sleep(3)                           # 3초 뒤에
driver.get("https://www.naver.com")     # 네이버로 이동
time.sleep(3)                           # 3초 뒤에

# 페이지 이동
driver.back()                           # 뒤로 이동
time.sleep(3)
driver.forward()                        # 앞으로 이동
time.sleep(3)
driver.refresh()                        # 새로 고침

# 크기 변경
driver.set_window_size(1200, 800)
time.sleep(3)

# 최대화
driver.maximize_window()
time.sleep(3)

# 최소화
driver.minimize_window()
time.sleep(3)

# 창 배치
driver.set_window_position(50, 50) # 왼쪽 위를 기준으로 50, 50을 이동한 위치
time.sleep(3)

# 브라우저 닫기
driver.quit()                           # 브라우저 종료
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

# Chrome 브라우저 실행
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

driver.get("https://www.selenium.dev/selenium/web/web-form.html")

# 웹 페이지의 요소 선택과 제어
text_input = driver.find_element(By.ID, "my-text-id")

time.sleep(3)

# 요소 제어
text_input.send_keys("Hello Selenium!")
time.sleep(3)
text_input.clear()
time.sleep(3)
text_input.send_keys("Coding")
time.sleep(3)

submit_btn = driver.find_element(By.TAG_NAME, "button")
submit_btn.click()

time.sleep(5)

driver.quit()
```

```python
# 구글 검색 예제
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

driver.get("https://www.google.com")

search_box = driver.find_element(By.NAME, "q")

search_box.clear()

search_box.send_keys("Python Selenium", Keys.RETURN)

time.sleep(10)

titles = driver.find_elements(By. TAG_NAME, "h3")[:5]

for t in titles:
    print(t.text)

time.sleep(5)

driver.quit()
```
